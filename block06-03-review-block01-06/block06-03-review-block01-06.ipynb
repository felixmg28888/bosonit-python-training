Crear un programa que tire un dado. Tras tener un resultado, pide al usuario que diga si quiere que se genere otro número. Nota: Para pedir información al usuario se puede utilizar *variable = input().

#### **Ejercicio 3**

Crear una función a la que le pasamos una lista y nos devuelve la misma lista pero ordenada al azar. Utilizar la función random.choice().

```python
import random


def ordenar_al_azar(lista):
    lista_ordenada_al_azar = []
    while lista:
        elemento = random.choice(lista)
        lista_ordenada_al_azar.append(elemento)
        lista.remove(elemento)
    return lista_ordenada_al_azar


lista1 = [1, 2, 3, 4, 5]
lista2 = ["a", "b", "c", "d", "e", "f", "g"]
lista3 = ["Invierno", (0, 0), 25, 0.2]

print("------------------------------------------------------------")
print("Lista1:", lista1)
print("Lista1_ordenada al azar", ordenar_al_azar(lista1))
print("------------------------------------------------------------")
print("Lista2:", lista2)
print("Lista1_ordenada al azar", ordenar_al_azar(lista2))
print("------------------------------------------------------------")
print("Lista3:", lista3)
print("Lista3_ordenada al azar", ordenar_al_azar(lista3))
print("------------------------------------------------------------")
```

#### **Ejercicio 4**

Crear una clase contraseña. Tendrá un método que nos diga si la contraseña es válida (usando el código del otro día). Será capaz de puntuar una contraseña (longitud + n_simbolos + n_numero) y comparar contraseñas entre sí. Al imprimir por pantalla una cadena mostraremos asteriscos.

```python
import re


class Contrasena:
    def __init__(self, contrasena):
        self.contrasena = contrasena

    def __str__(self):
        # Devuelve la contraseña con asteriscos
        return '*' * len(self.contrasena)

    def get_nombre_variable(self):
        for nombre, valor in globals().items():
            if valor is self:
                self.nombre_variable = nombre
                break

        return self.nombre_variable

    def comprobacion(self):
        # La contraseña debe tener al menos 8 caracteres
        if len(self.contrasena) < 8:
            return "longitud"

        # La contraseña debe contener al menos una letra minúscula, una mayúscula, un número y un carácter especial
        if not re.search(r'[a-z]', self.contrasena):
            return "minusculas"
        if not re.search(r'[A-Z]', self.contrasena):
            return "mayusculas"
        if not re.search(r'[0-9]', self.contrasena):
            return "numeros"
        if not re.search(r'[!@#$%^&*(),.?":{}|<>]', self.contrasena):
            return "simbolos"

        # Si ha pasado todas las comprobaciones, la contraseña es válida
        return "valida"

    def es_valida(self):
        comprobacion = self.comprobacion()
        if comprobacion == "longitud":
            return False, "La contraseña debe tener al menos 8 caracteres"
        elif comprobacion == "minusculas":
            return False, "La contraseña debe contener al menos una letra minúscula"
        elif comprobacion == "mayusculas":
            return False, "La contraseña debe contener al menos una letra mayúscula"
        elif comprobacion == "numeros":
            return False, "La contraseña debe contener al menos un número"
        elif comprobacion == "simbolos":
            return False, "La contraseña debe contener al menos un carácter especial"
        else:
            return True, "La contraseña es válida"

    def puntuar(self):
        es_valida = self.es_valida()
        if not es_valida[0]:
            return "Al no ser válida no se puede puntuar.", es_valida[1]

        # Longitud de la contraseña
        longitud = len(self.contrasena)

        # Número de símbolos (cualquier carácter que no sea letra ni número)
        n_simbolos = len(re.findall(
            r'[!@#$%^&*(),.?":{}|<>]', self.contrasena))

        # Número de dígitos numéricos
        n_numeros = len(re.findall(r'[0-9]', self.contrasena))

        # Puntuación total
        puntuacion = longitud + n_simbolos + n_numeros

        return "Longitud:", longitud, "+ n_simbolos:", n_simbolos, "+ n_numeros:", n_numeros, "= Puntuación:", puntuacion

    def comparar(self, otra_contrasena):
        es_valida = self.es_valida()
        if not es_valida[0]:
            return "La contraseña '{}' no es válida. Con lo que no se puede puntuar ni comparar".format(self.contrasena)

        es_valida = otra_contrasena.es_valida()
        if not es_valida[0]:
            return "La contraseña '{}' no es válida. Con lo que no se puede puntuar ni comparar".format(
                otra_contrasena.contrasena)

        puntuacion1 = self.puntuar()
        puntuacion2 = otra_contrasena.puntuar()

        if puntuacion1 > puntuacion2:
            return "La contraseña '{}'--> {} es más fuerte que la contraseña '{}'--> {}".format(
                self.get_nombre_variable(), self.puntuar()[7],
                otra_contrasena.get_nombre_variable(), otra_contrasena.puntuar()[7])
        elif puntuacion1 < puntuacion2:
            return "La contraseña '{}'--> {} es más fuerte que la contraseña '{}'--> {}".format(
                self.get_nombre_variable(), self.puntuar()[7],
                otra_contrasena.get_nombre_variable(), otra_contrasena.puntuar()[7])
        else:
            return "Ambas contraseñas tienen la misma puntuación"


# Crear dos contraseñas
while True:
    contrasena1 = Contrasena(input("Escribe una contraseña:"))
    if contrasena1.es_valida()[0]:
        print("Contraseña válida")
        break
    else:
        print(contrasena1.es_valida()[1])

while True:
    contrasena2 = Contrasena(input("Escribe otra contraseña:"))
    if contrasena2.es_valida()[0]:
        print("Contraseña válida")
        break
    else:
        print(contrasena2.es_valida()[1])

# Comprobar si son válidas
print("------------------------------------------------------------")
print(contrasena1.get_nombre_variable(),
      "¿es válida? -->", contrasena1.es_valida())
print(contrasena2.get_nombre_variable(),
      "¿es valida? -->", contrasena2.es_valida())
print("------------------------------------------------------------")

# Puntuar las contraseñas
print(contrasena1.puntuar())
print(contrasena2.puntuar())
print("------------------------------------------------------------")

# Comparar las contraseñas
print(contrasena1.comparar(contrasena2))
print("------------------------------------------------------------")

# Imprimir las contraseñas con asteriscos
print(contrasena1)
print(contrasena2)
```

#### **Ejercicio 5**

Vamos a realizar una pequeña simulación con números aleatorios. Vamos a simular como antes el lanzamiento de un dado. Trataremos de obtener de una forma experimental la probabilidad de obtener distintos resultados:

1. Calcular la probabilidad de obtener un 6 en un lanzamiento.
2. Calcular la probabilidad de obtener al menos cuatro 6 en 5 lanzamientos.
3. Calcular la probabilidad de obtener un número mayor en el segundo lanzamiento que en el primer lanzamiento.

##### **Teoría**

**1. Calcular la probabilidad de obtener un 6 en un lanzamiento.**<br>
Se lanza un dado un número suficientemente grande de veces y se registra cuántas veces sale un 6. Luego, se divide el número de veces que salió el 6 entre el número total de lanzamientos.

$$ P(6) = Número de veces que se obtuvo un 6 / Número total de lanzamientos$$

**2. Calcular la probabilidad de obtener al menos cuatro 6 en 5 lanzamientos.**<br>

Se realizan 5 lanzamientos y se registra cuántas veces sale el 6. Si el número de veces que sale el 6 es igual o mayor que cuatro, consideraremos que hemos obtenido al menos cuatro 6. Se repetirá este proceso un número suficientemente grande de veces y se dividirá el número de veces que seobtuvo al menos cuatro 6 entre el número total de repeticiones para obtener la probabilidad experimental.

$$P(al menos cuatro 6) = Número de veces que se obtuvieron al menos cuatro 6 / Número total de repeticiones$$


**3. Calcular la probabilidad de obtener un número mayor en el segundo lanzamiento que en el primer lanzamiento.**<br>
Se realizan una serie de lanzamientos del dado y se compararán los valores obtenidos en el primer lanzamiento y en el segundo lanzamiento. Se registra cuántas veces se obtiene un número mayor en el segundo lanzamiento que en el primer lanzamiento. Se repite este proceso un número suficientemente grande de veces y se dividirá el número de veces que se obitene un número mayor en el segundo lanzamiento entre el número total de repeticiones para obtener la probabilidad experimental.

$$P(segundo lanzamiento > primer lanzamiento) = Número de veces que el segundo lanzamiento es mayor / Número total de repeticiones$$

```python
import random

# Función para simular el lanzamiento de un dado


def tirar_dado():
    return random.randint(1, 6)

# 1) Calcular la probabilidad de obtener un 6 en un lanzamiento


def calcular_probabilidad_un_seis(num_lanzamientos):
    lanzamientos_con_seis = 0

    for _ in range(num_lanzamientos):
        resultado = tirar_dado()
        if resultado == 6:
            lanzamientos_con_seis += 1

    probabilidad = lanzamientos_con_seis / num_lanzamientos
    return probabilidad

# 2) Calcular la probabilidad de obtener al menos cuatro 6 en 5 lanzamientos


def calcular_probabilidad_cuatro_seis(num_repeticiones):
    contador_exitos = 0

    for _ in range(num_repeticiones):
        seis_obtenidos = 0

        for _ in range(5):
            resultado = tirar_dado()
            if resultado == 6:
                seis_obtenidos += 1

        if seis_obtenidos >= 4:
            contador_exitos += 1

    probabilidad = contador_exitos / num_repeticiones
    return probabilidad

# 3) Calcular la probabilidad de obtener un número mayor en el segundo lanzamiento que en el primer lanzamiento


def calcular_probabilidad_segundo_mayor(num_repeticiones):
    contador_exitos = 0

    for _ in range(num_repeticiones):
        primer_lanzamiento = tirar_dado()
        segundo_lanzamiento = tirar_dado()
        if segundo_lanzamiento > primer_lanzamiento:
            contador_exitos += 1

    probabilidad = contador_exitos / num_repeticiones
    return probabilidad


# Ejemplo
num_repeticiones = 10000
prob_un_seis = calcular_probabilidad_un_seis(num_repeticiones)
prob_cuatro_seis = calcular_probabilidad_cuatro_seis(num_repeticiones)
prob_segundo_mayor = calcular_probabilidad_segundo_mayor(num_repeticiones)

print("Probabilidad de obtener un 6 en un lanzamiento:", prob_un_seis)
print("Probabilidad de obtener al menos cuatro 6 en 5 lanzamientos:", prob_cuatro_seis)
print("Probabilidad de obtener un número mayor en el segundo lanzamiento que en el primer lanzamiento:", prob_segundo_mayor)
```

#### **Ejercicio 6**

Tenemos el siguiente juego: pagamos 1 euro y podemos tirar 4 dados. Si la suma de los 4 dados es menor que 9 nos pagan _r_ euros, en caso contrario perdemos el euro apostado (supondremos r=10). ¿Jugarías?

##### **Teoría**

Para determinar si sería beneficioso jugar o no a este juego, necesitamos calcular la expectativa matemática o valor esperado.

El valor esperado se calcula multiplicando cada posible resultado por su probabilidad y sumando estos productos. Si el valor esperado es mayor que cero, entonces se espera un beneficio a largo plazo, mientras que si es menor que cero, se espera una pérdida a largo plazo.

En este caso, si la suma de los 4 dados es menor que 9, ganamos 10 euros (r = 10), y si la suma es 9 o mayor, perdemos 1 euro. La probabilidad de obtener una determinada suma se puede calcular enumerando todas las combinaciones posibles de los 4 dados y contando cuántas sumas cumplen la condición requerida.

```python
import random


def tirar_dados():
    return [random.randint(1, 6) for _ in range(4)]


def calcular_prob_menor9():
    suma_menor_nueve = 0
    total_combinaciones = 6 ** 4  # 6 opciones posibles para cada dado, 4 dados en total

    for _ in range(total_combinaciones):
        dados = tirar_dados()
        suma = sum(dados)
        if suma < 9:
            suma_menor_nueve += 1

    probabilidad_menor_nueve = suma_menor_nueve / total_combinaciones
    return probabilidad_menor_nueve


def calcular_prob_mayor9():
    suma_nueve_o_mayor = 0
    total_combinaciones = 6 ** 4  # 6 opciones posibles para cada dado, 4 dados en total

    for _ in range(total_combinaciones):
        dados = tirar_dados()
        suma = sum(dados)
        if suma >= 9:
            suma_nueve_o_mayor += 1

    probabilidad_nueve_o_mayor = suma_nueve_o_mayor / total_combinaciones
    return probabilidad_nueve_o_mayor


def calcular_valor_esperado():
    ganancia = 10
    costo_apuesta = 1

    probabilidad_menor_nueve = calcular_prob_menor9()
    probabilidad_nueve_o_mayor = calcular_prob_mayor9()

    valor_esperado = (probabilidad_menor_nueve * ganancia) - \
        (probabilidad_nueve_o_mayor * costo_apuesta)

    return valor_esperado


# Ejemplo

probabilidad_menor_nueve = calcular_prob_menor9()
probabilidad_nueve_o_mayor = calcular_prob_mayor9()
valor_esperado = calcular_valor_esperado()

print("Probabilidad de obtener una suma menor a 9:", probabilidad_menor_nueve)
print("Probabilidad de obtener una suma de 9 o mayor:", probabilidad_nueve_o_mayor)
print("Valor esperado:", valor_esperado)

if valor_esperado > 0:
    print("¿Jugarías?...Sí")
else:
    print("¿Jugarías?...No")
```
